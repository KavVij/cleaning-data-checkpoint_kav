## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
#Finding missing values
total_cells = np.product(df.shape)

missing_cells = pd.isnull(df).sum()

total_missing = missing_cells.sum()

percentage_missing = round((total_missing/total_cells), 2) *100

print(percentage_missing, "% Missing cells from the data") 
```

```python
#to find the missing percentages
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100,2)))
```

```python
#Creating a new database called icecream. Adding a new table to database called icecrea_table using the data in the 
#dataframe.

con=sl.connect('icecream', timeout=1)
con.execute('''Drop table icecream_table''')
df.to_sql('icecream_table',con)
con.commit()
```

```python
#Finding icecreams with rating_count greater than the mean(78),to remove the one which was reviewed by less than the mean 
#number of people(less rating_count)
with con:
    data=con.execute('select* from icecream_table where rating_count>=78')
    for row in data:
        print(row)
```

```python
#Finding the 5 highest rating ice creams 
with con:
     data = con.execute('''select brand,name,rating,rating_count
                     from icecream_table
                     where rating_count>=78
                     order by rating DESC LIMIT 5''')
for row in data:
        print(row)
                  
```

```python
#Finding the 5 lowest rating ice cream
with con:
     data = con.execute('''select brand,name,rating,rating_count
                     from icecream_table
                     where rating_count>=78
                     order by rating limit 5''')
for row in data:
        print(row)
```
